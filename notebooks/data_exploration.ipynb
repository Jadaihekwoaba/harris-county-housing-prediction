```python
# Load building details
print("Loading building data...")
df_building = pd.read_csv('/Users/jadaihekwoaba/Desktop/housing-prediction/data/raw/Real_building_land/building_res.txt',
                          sep='\t',
                          encoding='latin1',
                          low_memory=False,
                          on_bad_lines='skip')

print(f"Total building records: {len(df_building)}")

# Show columns
print("Building columns:")
print(df_building.columns.tolist())
```

```python
# Filter building data to only accounts in your residential dataset
residential_accounts = df_residential['acct'].unique()
df_building_filtered = df_building[df_building['acct'].isin(residential_accounts)]

print(f"Building records for Oak Forest: {len(df_building_filtered)}")

# Merge the two datasets
df_merged = df_residential.merge(df_building_filtered, on='acct', how='left')

print(f"Merged dataset: {len(df_merged)} rows")

# Look at key columns
df_merged[['acct', 'site_addr_1', 'site_addr_3', 'tot_appr_val', 'im_sq_ft', 'date_erected', 'Neighborhood_Code']].head(10)
```

```python
# Basic statistics
print("=== DATA SUMMARY ===")
print(f"\nTotal properties: {len(df_merged)}")

print("\n=== PROPERTY VALUES ===")
print(df_merged['tot_appr_val'].describe())

print("\n=== SQUARE FOOTAGE ===")
print(df_merged['im_sq_ft'].describe())

print("\n=== YEAR BUILT ===")
print(df_merged['date_erected'].describe())

# Check for missing data
print("\n=== MISSING DATA ===")
print(df_merged[['tot_appr_val', 'im_sq_ft', 'date_erected']].isnull().sum())

# Check for zeros (properties with no value or sqft)
print("\n=== ZERO VALUES ===")
print(f"Properties with $0 value: {(df_merged['tot_appr_val'] == 0).sum()}")
print(f"Properties with 0 sqft: {(df_merged['im_sq_ft'] == 0).sum()}")
```

```python
# Make a copy to clean
df_clean = df_merged.copy()

print(f"Starting with: {len(df_clean)} properties")

# Remove properties with missing critical data
df_clean = df_clean.dropna(subset=['tot_appr_val', 'im_sq_ft', 'date_erected'])
print(f"After removing missing data: {len(df_clean)}")

# Remove properties with 0 or unreasonable sqft
df_clean = df_clean[(df_clean['im_sq_ft'] > 300) & (df_clean['im_sq_ft'] < 10000)]
print(f"After removing bad sqft: {len(df_clean)}")

# Remove properties with unreasonable values (likely errors or commercial)
df_clean = df_clean[(df_clean['tot_appr_val'] > 50000) & (df_clean['tot_appr_val'] < 3000000)]
print(f"After removing outlier values: {len(df_clean)}")

# Create additional features
df_clean['property_age'] = 2025 - df_clean['date_erected']
df_clean['price_per_sqft'] = df_clean['tot_appr_val'] / df_clean['im_sq_ft']

print(f"\n=== FINAL CLEAN DATASET ===")
print(f"Total properties: {len(df_clean)}")
print(f"\nPrice per sqft stats:")
print(df_clean['price_per_sqft'].describe())

# Save the clean data
df_clean.to_csv('/Users/jadaihekwoaba/Desktop/housing-prediction/data/processed/oak_forest_clean.csv', 
                index=False)
print("\Saved clean data")
```
